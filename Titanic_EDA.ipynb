```python
import numpy as np

sample_means = []
for _ in range(1000):
    sample = df['Age'].dropna().sample(30)
    sample_means.append(sample.mean())

sns.histplot(sample_means, kde=True)
plt.title("CLT Demonstration")
plt.show()
```
